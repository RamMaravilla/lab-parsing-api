## Reading the data

In the next cell, use Python to obtain data from the API endpoint you chose in the previous step. Workflow:

```python
# your code here

#Import libraries
import json
import requests
import pandas as pd
from pandas.io.json import json_normalize
```

```python
#Define the initial API and endpoint URL

url = 'https://rickandmortyapi.com/api/character/'   #guarda la url

response = requests.get('https://rickandmortyapi.com/api/character/')  #de aquí se va a bajar la información
results = response.json()  #aqui se tiene la respuesta bajada
#results
```

```python
#Contar las páginas de personjes
count_pages = results['info']['pages']  #cuenta lás páginas par saber cómo vamos
count_pages
```

```python
#Describir la primera página para saber por dónde le vamos a entrar
# Info de la primra pagina
data_character = json_normalize(results['results'])
data_character
```

```python
#Observamos que la información ya se descargó en un panda leible. Veamos si se puede hacer con flattened.
#la flatenilización no funcionó. Lo dejamos como está por el momento
#flattened_data = json_normalize(results)
#lattened_data

#flattened_data1 = json_normalize(flattened_data.articles[0])
#flattened_data1
```

```python
# Lo hacemos para todas las páginas de Rick&Morty

for page in range(2,count_pages+1):
    url_character = 'https://rickandmortyapi.com/api/character/?page='+str(page)
    print(url_character)
    character_json = requests.get(url_character).json()  #obtiene la info en formato json
    flattened_data = json_normalize(character_json['results']) #la hace flat para hacerla leible
    data_character = pd.concat([data_character,flattened_data]) #junta todas las paginas de data_character(página 1)
                                                                #con toda la información de flattened data (p 2:)
    
data_character
```
